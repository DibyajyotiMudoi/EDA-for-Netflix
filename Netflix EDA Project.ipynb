```python
#Distribution of movives and tv shows 
plt.figure(figsize=(12,6))
netflix['type'].value_counts().plot(kind='bar', color=['red','skyblue'])
plt.title('Distribution of Movies and TV shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# Country-wise analysis
plt.figure(figsize=(16, 8))
sns.countplot(x='country', data=netflix, order=netflix['country'].value_counts().index[:15], palette='viridis')
plt.title('Top 15 Countries Producing Content on Netflix')
plt.xticks(rotation=45)
plt.show()
```

```python
# Rating analysis
plt.figure(figsize=(10, 6))
sns.countplot(x='rating', data=netflix, palette='Pastel1')
plt.title('Distribution of Content Ratings on Netflix')
plt.show()
```
